<b>3. Create a CSV file called books.csv by using these lines:
title,author,year
The Weirdstone of Brisingamen,Alan Garner,1960
Perdido Street Station,China Miéville,2000
Thud!,Terry Pratchett,2005
The Spellman Files,Lisa Lutz,2007
Small Gods,Terry Pratchett,1992

```python
rows = [['title', 'author', 'year'],
['The Weirdstone of Brisingamen', 'Alan Garner', 1960],
['Perdido Street Station', 'China Miéville', 2000],
['Thud!', 'Terry Pratchett', 2005],
['The Spellman Files', 'Lisa Lutz', 2007],
['Small Gods', 'Terry Pratchett', 1992]]

import csv
with open('books.csv', 'w') as f:
    csv_writer = csv.writer(f, delimiter=',', lineterminator='\n')
    csv_writer.writerows(rows)
```

<b>4. Use the sqlite3 module to create a SQLite database called books.db, and a table called books with
these fields: title (text), author (text), and year (integer).

```python
import sqlite3

conn = sqlite3.connect('books.db')
cur = conn.cursor()

cur.execute('DROP TABLE IF EXISTS books')
conn.commit()

query = '''CREATE TABLE IF NOT EXISTS books(title text,
            author text,
            year integer)'''

cur.execute(query)
conn.commit()
```

<b>5. Read books.csv and insert its data into the book table.

```python
with open('books.csv', 'r') as f:
    csv_reader = csv.reader(f)
    for n, row in enumerate(csv_reader):
        if n != 0:
            cur.execute(f'INSERT INTO books VALUES("{row[0]}", "{row[1]}", {row[2]})')
            
conn.commit()
```

<b>6. Select and print the title column from the book table in alphabetical order.

```python
cur.execute('SELECT title FROM books ORDER BY title')
for res in cur.fetchall():
    print(res[0])
```

<b>7. From the book table, select and print all columns in the order of publication.

```python
cur.execute('SELECT * FROM books ORDER BY year')
for res in cur.fetchall():
    print(res)
```
